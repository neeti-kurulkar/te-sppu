```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the Titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
# Display the first few rows to understand the structure of the dataset
print("\nFirst 5 Rows of Titanic Dataset:")
print(titanic.head())
```

```python
# Display column names
print("\nColumn Names in Titanic Dataset:")
print(titanic.columns)
```

```python
# Display summary statistics
print("\nSummary Statistics of Titanic Dataset:")
print(titanic.describe())
```

```python
# Countplot of passengers by class
plt.figure(figsize=(8, 5))
sns.countplot(x='class', data=titanic, palette='coolwarm')
plt.title("Number of Passengers in Each Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```

```python
# Survival rate by class
plt.figure(figsize=(8, 5))
sns.barplot(x='class', y='survived', data=titanic, palette='coolwarm')
plt.title("Survival Rate by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

```python
# Age distribution by class
plt.figure(figsize=(8, 5))
sns.boxplot(x='class', y='age', data=titanic, palette='coolwarm')
plt.title("Age Distribution by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
```

```python
# Count of male and female passengers
plt.figure(figsize=(8, 5))
sns.violinplot(x='sex', data=titanic, palette='coolwarm')
plt.title("Number of Male and Female Passengers")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# Survival rate by gender
plt.figure(figsize=(8, 5))
sns.barplot(x='sex', y='survived', data=titanic, palette='coolwarm')
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.show()
```

```python
### Visualizing Ticket Price Distribution**

plt.figure(figsize=(10, 6))
sns.histplot(titanic['fare'], bins=30, kde=True, color='blue')
plt.title("Distribution of Ticket Prices (Fare)")
plt.xlabel("Fare Price")
plt.ylabel("Number of Passengers")
plt.show()
```
