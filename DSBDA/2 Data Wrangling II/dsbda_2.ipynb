The dataset "marksheet.csv" containins information about students, including their ID, name, gender, age, section, and scores in various subjects (Science, English, History, and Maths).
id: A unique identifier for each student (integer).
Name: The name of the student (string).
Gender: The gender of the student (categorical, e.g., Male, Female).
Age: The age of the student (integer).
Section: The section in which the student belongs (categorical, e.g., A, B, C).
Science: The score the student received in the Science subject (integer, could have missing values).
English: The score the student received in the English subject (integer, could have missing values).
History: The score the student received in the History subject (integer, could have missing values).
Maths: The score the student received in the Maths subject (integer, could have missing values).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read the csv file
df = pd.read_csv("student_data.csv")
```

```python
df
```

```python
# check for missing values
df.isnull().sum()
```

```python
# Handling missing values
# For categorical columns fill with mode
df['Section'] = df['Section'].fillna(df['Section'].mode()[0])
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])

# For numerical columns fill with median
df['Science'] = df['Science'].fillna(df['Science'].median())
df['English'] = df['English'].fillna(df['English'].median())
df['History'] = df['History'].fillna(df['History'].median())
df['Maths'] = df['Maths'].fillna(df['Maths'].median())
```

```python
df
```

```python
# Select subject columns for boxplot
subjects = ['Science', 'English', 'History', 'Maths']

# Create a boxplot for each subject
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[subjects])
plt.title('Boxplot for Outlier Detection in Subjects')
plt.ylabel('Scores')
plt.xlabel('Subjects')
plt.show()
```

```python
def detect_outliers_value_range(df, column, lower_limit=0, upper_limit=100):
    outliers = df[(df[column] < lower_limit) | (df[column] > upper_limit)]
    return outliers
```

```python
# Detect outliers in Science
outliers_science = detect_outliers_value_range(df, 'Science')
print("Outliers in Science:\n", outliers_science)

# Detect outliers in English
outliers_english = detect_outliers_value_range(df, 'English')
print("\nOutliers in English:\n", outliers_english)

# Detect outliers in History
outliers_maths = detect_outliers_value_range(df, 'History')
print("\nOutliers in History:\n", outliers_maths)

# Detect outliers in Maths
outliers_maths = detect_outliers_value_range(df, 'Maths')
print("\nOutliers in Maths:\n", outliers_maths)
```

```python
# replace outliers with median values
def replace_outliers_with_median(df, column, lower_limit=0, upper_limit=100):
    median_value = df[column].median()
    df[column] = df[column].apply(lambda x: median_value if x < lower_limit or x > upper_limit else x)
    return df

df = replace_outliers_with_median(df, 'Science')
df = replace_outliers_with_median(df, 'English')
df = replace_outliers_with_median(df, 'History')
df = replace_outliers_with_median(df, 'Maths')
```

```python
df
```

```python
# Check skewness before transformation
before_skewness = df['Science'].skew()

# Apply log transformation (only on positive values)
df['Science_log'] = df['Science'].apply(lambda x: np.log(x+1) if x > 0 else 0)

# Check skewness after transformation
after_skewness = df['Science_log'].skew()

# Plot the distribution of 'Science' before and after transformation
plt.figure(figsize=(12, 6))

# Original 'Science' variable
plt.subplot(1, 2, 1)
sns.histplot(df['Science'], kde=True)
plt.title('Distribution of Science (Before Transformation)')
plt.xlabel('Science Scores')
plt.ylabel('Frequency')

# Log-transformed 'Science' variable
plt.subplot(1, 2, 2)
sns.histplot(df['Science_log'], kde=True)
plt.title('Distribution of Science (After Log Transformation)')
plt.xlabel('Log-transformed Science Scores')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
print("Skewness before transformation:", before_skewness)
print("Skewness after transformation:", after_skewness)
```
